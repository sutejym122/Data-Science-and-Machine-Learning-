Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
from matplotlib import pyplot as plt
Days=[1,2,3,4,5,6,7,8]
Speed=[20,40,45,60,83,75,120,45]
plt.plot(Days,Speed)
plt.xlabel('Days')
plt.ylabel('Speed')
plt.title('car speed measurement')
plt.show()
```

Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.

```python
plt.plot(Days,Speed,'g--',marker='+',markersize=20,mfc='red',color='b')
plt.xlabel('Days')
plt.ylabel('Speed')
plt.title('Car speed measurement ')
plt.show()
```

Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

days=[1,2,3,4,5,6,7,8]

```python
days=[1,2,3,4,5,6,7,8]

max_speed=[80,91,92,88,77,79,76,75]

min_speed=[42,43,40,42,33,36,34,35]

avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,max_speed,'r',label="maximum speed")
plt.plot(days,min_speed,'g',label="minimum speed")
plt.plot(days,avg_speed,'b',label="average speed")
plt.xlabel('days')
plt.ylabel('speed')
plt.legend()
plt.grid()
plt.axes()
plt.show()
```

Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.show()
```

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.
