```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn 
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/50_Startups.csv")
```

```python
df
#first 4 columns is X Profit is Y 
```

```python
y=df['Profit']
x=df.drop(['Profit','State'],axis=1) # not able to convert string to float (in State column) s State is dropped

print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=5)
```

```python
lm=LinearRegression()
lm.fit(xtrain,ytrain)
```

```python
ypred_train=lm.predict(xtrain)
ypred_test=lm.predict(xtest)
```

```python
df=pd.DataFrame(ypred_test,ytest)
df.head()
```

```python
df=pd.DataFrame(ypred_train,ytrain)
df.head()
```

```python
from sklearn.metrics import  mean_squared_error
mse=mean_squared_error(ypred_test,ytest)
print(mse)
```
